## Pre-Processing Data

in this part, a dataset gets generated which is ready for matching with the AP data and the correlation calculation.

#### Creating Emergency Admissions Data and exploring:

```python
# Defining emergency columns
emergency_columns = [col for col in df_health.columns if 'Emergency' in col]
emergency_columns
```

```python
columns_to_include = ['dates', 'Name'] + emergency_columns
df_emergency = df_health[columns_to_include]
df_emergency.info()
```

```python
df_emergency.head(5)
```

```python
df_emergency.describe()
```

```python
# Plot
grouped_df = df_emergency.groupby('Name')

plt.figure(figsize=(10, 6))  
for name, group in grouped_df:
    plt.plot(group['dates'], group['Emergency Admissions Total Emergency Admissions'], label=name)

plt.xlabel('Date')
plt.ylabel('Total Emergency Admissions')
plt.title('Total Emergency Admissions Over Time')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

Based on the exploring and discussions over the data, values in **"Emergency Admissions Total Emergency Admissions"** column is valuable for this project. So the focus will be only on this column in furthur steps.  
    
Emergency admissions counts are provided for 4 hospital and health centres during the period 2015 to 2024.  
It seems that "University Hospitals Bristol NHS Foundation Trust" and "University Hospitals Bristol And Weston NHS Foundation Trust" are the same health centres but the name has changed due to unknown reasons for us.    
   
By the way, the counts of emergency admissions in whole Bristol city is what that matters so a dataset should be generated that has the **SUM** value of all these Bristol health centres.  


#### Generating data with SUM value of Emergency admissions values:

```python
df_grouped = df_emergency.groupby('dates')['Emergency Admissions Total Emergency Admissions'].sum()
df_grouped = df_grouped.reset_index()
```

```python
df_grouped = df_grouped.rename(columns={'dates': 'Date', 
                                        'Emergency Admissions Total Emergency Admissions': 'Total Emergency Admissions'})
```

```python
# Generating csv file
df_grouped.to_csv('Health data - Final.csv', index=False)
```

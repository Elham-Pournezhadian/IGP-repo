```python
import pandas as pd

# Assuming 'data' contains your dataset and the dates are stored in a column named 'dates'

# Convert the 'dates' column to datetime format
data['dates'] = pd.to_datetime(data['dates'], format='%B %Y')

# Display the updated DataFrame
print(data['dates'])
```

```python
data['year'] = data['dates'].dt.year

# Display the updated DataFrame
print(data['year'])
```

```python


for column in data.columns[3:-2]:
    plt.figure(figsize=(15, 10))  # Adjust width and height as needed
    # Create line plot
    ax = sns.lineplot(data=data, x='dates', y=column, hue='Name')

    plt.xticks(rotation=90)  # Rotate the labels for better visibility
    
    ax.set_title(column + ' over time')
   


    # Show plot
    plt.show()
```
